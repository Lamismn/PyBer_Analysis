```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, marker='d', color="green", linewidth=1.5, label='Boston')
plt.xlabel('Date')
plt.ylabel('Fare')
plt.ylim(0,45)
plt.title('PyBer Fare by Month')
plt.legend()
plt.grid()
```

```python
plt.barh(x_axis, y_axis, color='magenta')
plt.xlabel('Date')
plt.ylabel('Fare')
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis, color='cadetblue', marker='d' )
```

```python
plt.scatter(x_axis, y_axis, s=[ i*5 for i in y_axis], color='darksalmon', edgecolor='maroon')
```

```python
categories = ['pants', 'shirts', 'outerwear', 'pajamas', 'underwear', 'shoes']
revenue = [25, 15, 5, 20, 10, 25]
explode = [0, 1, 0, 0, 0, 0]
colors = ['slategray', 'lightsalmon', 'mediumspringgreen', 'cadetblue', 'tomato', 'deepskyblue']
plt.pie(revenue, labels=categories, explode=explode, colors=colors, autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
plt.pie(y_axis, labels=x_axis, autopct='%1.2f%%')
plt.show()
```
